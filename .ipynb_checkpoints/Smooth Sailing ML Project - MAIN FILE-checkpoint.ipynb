## Full Data Preprocessing & Train-Test split

```python
# Copy of data for visualisations
data_vis = data.copy()
# Save labels
y = data['activity']
# Drop user_snippets and labels
x = data.drop(['user_snippet', 'activity'], axis=1)
# Train-Test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

print('Training data size : ', X_train.shape)
print('Test data size : ', X_test.shape)
```

```python
# Save unique labels names
labels = ['Jogging', 'Walking','Standing','Downstairs','Upstairs','Sitting']
```

## Modelling

```python
# Function used for models' evaluations
def evaluate(model, X_train, y_train, X_test, y_test, labels):
    print('processing...\n')
    # Fit the data
    model.fit(X_train, y_train)  
    # Save predictions
    pred = model.predict(X_test)
    # Calculate accuracy score
    accuracy = metrics.accuracy_score(y_true=y_test, y_pred=pred)
    # Display accuracy, best fit parameters and best cross validation score
    print('Accuracy:\n {}\n'.format(accuracy))
    print('Best Parameters:\n{}\n'.format(model.best_params_))
    print('Best Cross Validation score:\n{}\n'.format(model.best_score_))
    
    # Show confusion matrix
    print('Confusion Matrix:')
    cm = metrics.confusion_matrix(y_test, pred)
    cm_plot = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    fig, ax = plt.subplots(figsize=(12,12))
    cm_plot.plot(ax=ax)
    return model
    
```

### Kaggle data

```python
# Load kaggle data and perform same operations as before
kaggle = pd.read_csv('Data/metadata/metadata_kaggle.csv')
kaggle = kaggle.drop(['x-axis__length', 'y-axis__length', 'z-axis__length'],axis=1)

signals_kaggle = pd.read_csv('Data/signals/signals_kaggle.csv')
signals_kaggle['x-diffs'] = signals_kaggle['x-axis'].diff()
signals_kaggle['y-diffs'] = signals_kaggle['y-axis'].diff()
signals_kaggle['z-diffs'] = signals_kaggle['z-axis'].diff()
signals_kaggle = signals_kaggle[signals_kaggle['timestamp'] != 0]

means_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].mean().rename(columns= {'x-diffs':'x-diff_mean', 'y-diffs':'y-diff_mean', 'z-diffs':'z-diff_mean'})
medians_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].median().rename(columns= {'x-diffs':'x-diff_median', 'y-diffs':'y-diff_median', 'z-diffs':'z-diff_median'})
maxes_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].max().rename(columns= {'x-diffs':'x-diff_max', 'y-diffs':'y-diff_max', 'z-diffs':'z-diff_max'})
mins_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].min().rename(columns= {'x-diffs':'x-diff_min', 'y-diffs':'y-diff_min', 'z-diffs':'z-diff_min'})
stds_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].std().rename(columns= {'x-diffs':'x-diff_std', 'y-diffs':'y-diff_std', 'z-diffs':'z-diff_std'})
vars_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].var().rename(columns= {'x-diffs':'x-diff_var', 'y-diffs':'y-diff_var', 'z-diffs':'z-diff_var'})
sums_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].sum().rename(columns= {'x-diffs':'x-diff_sum', 'y-diffs':'y-diff_sum', 'z-diffs':'z-diff_sum'})
sems_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].sem().rename(columns= {'x-diffs':'x-diff_sem', 'y-diffs':'y-diff_sem', 'z-diffs':'z-diff_sem'})
skews_kaggle = signals_kaggle.groupby(['user_snippet'])[['x-diffs', 'y-diffs', 'z-diffs']].skew().rename(columns= {'x-diffs':'x-diff_skew', 'y-diffs':'y-diff_skew', 'z-diffs':'z-diff_skew'})
sems_kaggle_o = signals_kaggle.groupby(['user_snippet'])[['x-axis', 'y-axis', 'z-axis']].sem().rename(columns= {'x-axis':'x_sem', 'y-axis':'y_sem', 'z-axis':'z_sem'})
skews_kaggle_o = signals_kaggle.groupby(['user_snippet'])[['x-axis', 'y-axis', 'z-axis']].skew().rename(columns= {'x-axis':'x_skew', 'y-axis':'y_skew', 'z-axis':'z_skew'})

new_features_kaggle = pd.concat([means_kaggle, medians_kaggle, maxes_kaggle, mins_kaggle, stds_kaggle, vars_kaggle, sums_kaggle,
                                sems_kaggle, skews_kaggle, sems_kaggle_o, skews_kaggle_o], axis=1)

new_features_kaggle['x-diff_skew'].fillna(new_features_kaggle['x-diff_skew'].mean(), inplace=True)
new_features_kaggle['y-diff_skew'].fillna(new_features_kaggle['y-diff_skew'].mean(), inplace=True)
new_features_kaggle['z-diff_skew'].fillna(new_features_kaggle['z-diff_skew'].mean(), inplace=True)
new_features_kaggle['x_skew'].fillna(new_features_kaggle['x_skew'].mean(), inplace=True)
new_features_kaggle['y_skew'].fillna(new_features_kaggle['y_skew'].mean(), inplace=True)
new_features_kaggle['z_skew'].fillna(new_features_kaggle['z_skew'].mean(), inplace=True)

kaggle = kaggle.merge(new_features_kaggle, how='left', on='user_snippet')

kaggle_ids = kaggle.user_snippet
kaggle = kaggle.drop('user_snippet', axis=1)

kaggle
```
