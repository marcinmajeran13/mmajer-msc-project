```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=100)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import metrics
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier

labels = ['LAYING', 'SITTING', 'STANDING', 'WALKING', 'WALKING_DOWNSTAIRS',
       'WALKING_UPSTAIRS']
```

```python
def evaluate(model, X_train, y_train, X_test, y_test, labels):
    print('processing...\n')
    # Fit the data
    model.fit(X_train, y_train)  
    # Save predictions
    pred = model.predict(X_test)
    # Calculate accuracy score
    accuracy = metrics.accuracy_score(y_true=y_test, y_pred=pred)
    # Display accuracy, best fit parameters and best cross validation score
    print('Accuracy:\n {}\n'.format(accuracy))
    print('Best Parameters:\n{}\n'.format(model.best_params_))
    print('Best Cross Validation score:\n{}\n'.format(model.best_score_))
    
    # Show confusion matrix
    print('Confusion Matrix:')
    cm = metrics.confusion_matrix(y_test, pred)
    cm_plot = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    fig, ax = plt.subplots(figsize=(12,12))
    cm_plot.plot(ax=ax)
    return model
    
```

```python
rf = RandomForestClassifier()
param_grid = {
#     'max_depth': [7],
#     'criterion' : ['entropy'],
#     'max_features': [25],
#     'min_weight_fraction_leaf': [0.15],
#     'n_estimators': [300]
    'random_state': [1,2,3,4,5,6,7,8,9]
}
rf_grid = GridSearchCV(rf, param_grid, n_jobs=-1)
rf_model = evaluate(rf_grid, X_train, y_train, X_test, y_test, labels)
```
